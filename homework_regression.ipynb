```python
#パッケージをインポート
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#パッケージのモジュールをインポート
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
#matplotlibの便利なコマンド
%matplotlib inline
```

```python
#クラス・メソッド
#index_col=0 Unnamed: 0をなくす
df= pd.read_csv('california_housing_cleansing.csv',index_col=0)
df
```

```python
#変数を作成
#列に沿った処理はaxis=0、行に沿った処理はaxis=1
#以下わからない
#X = df.drop("所得", axis=1)
#y = df["所得"]

X = df.drop(columns=['所得'])
y = df['所得']
```

```python
X
```

```python
y
```

```python
#インスタンス化　スケールを合わせる（平均0　標準偏差1）
sc = StandardScaler()
```

```python
#スケール変更
Xnorm = sc.fit_transform(X)
```

```python
#test_size,random_stateの設定
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
#インスタンス化　変数の値を予測
reg = LinearRegression()
```

```python
#学習する
reg.fit(X_train, y_train)
```

```python
#予測する
y_pred = reg.predict(X_test)
y_pred
```

```python
# y_predの正解率を評価する
#
r2_score(y_test, y_pred)
```
